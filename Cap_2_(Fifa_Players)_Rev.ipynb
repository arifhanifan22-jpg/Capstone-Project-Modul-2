```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data FIFA Players dari Drive
df_fifa = pd.read_csv('/content/drive/MyDrive/CAPSTONE 2/fifa_players.csv')

# Menampilkan data teratas
df_fifa.head(100)
```

```python
# A. Membuat fitur rasio harga per poin rating
df_fifa['value_per_ovr'] = df_fifa['value_euro'] / df_fifa['overall_rating']

# B. Membuat fungsi untuk mengkategorikan usia
def membuat_kategori_usia(umur):
    if umur <= 21:
        return 'Wonderkid (<=21)'
    elif umur <= 29:
        return 'Golden Age (22-29)'
    else:
        return 'Veteran (>=30)'

# Menerapkan fungsi ke dalam kolom baru
df_fifa['age_category'] = df_fifa['age'].apply(membuat_kategori_usia)

# C. Hitung ulang fitur growth yang sebelumnya
df_fifa['growth'] = df_fifa['potential'] - df_fifa['overall_rating']
```

```python
# --- INSIGHT 3: AKSELERASI KEMATANGAN RATING PEMAIN ---

# 1. Pastikan kolom growth_room sudah ada (selisih potential - overall_rating)
df_fifa['growth_room'] = df_fifa['potential'] - df_fifa['overall_rating']

# 2. Hitung Skor Akselerasi Pertumbuhan
# Rumus: growth_room dibagi dengan usia pemain.
# Semakin muda usia dan semakin besar sisa potensinya, skornya akan semakin tinggi.
df_fifa['acceleration_score'] = df_fifa['growth_room'] / df_fifa['age']

# 3. Filter pemain muda (misal usia <= 23) yang memiliki pertumbuhan nyata (growth_room > 0)
df_akselerasi = df_fifa[(df_fifa['age'] <= 23) & (df_fifa['growth_room'] > 0)].copy()

# 4. Tampilkan 10 besar pemain dengan akselerasi kematangan tertinggi
print("=== TOP 10 PEMAIN DENGAN AKSELERASI KEMATANGAN TERTINGGI ===")
print(df_akselerasi[['name', 'age', 'overall_rating', 'potential', 'growth_room', 'acceleration_score']]
      .sort_values(by='acceleration_score', ascending=False)
      .head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

# 1. Ambil top 10 data akselerasi tertinggi
top_10_akselerasi = df_akselerasi.sort_values(by='acceleration_score', ascending=False).head(10)

# 2. Atur ukuran kanvas grafik
plt.figure(figsize=(10, 6))

# 3. Buat Grafik Batang Horizontal
sns.barplot(
    x='acceleration_score',
    y='name',
    hue='name',
    data=top_10_akselerasi,
    palette='magma', # Menggunakan palet magma yang elegan
    legend=False
)

# 4. Tambahkan Judul dan Label
plt.title('Top 10 Pemain dengan Skor Akselerasi Kematangan Tertinggi', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Skor Akselerasi (Growth Room / Usia)', fontsize=12)
plt.ylabel('Nama Pemain', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# --- PERBAIKAN INSIGHT 4: STRUKTUR GAJI BERDASARKAN POSISI BERMAIN ---

# 1. Kelompokkan data berdasarkan Posisi Bermain (positions)
df_posisi = df_fifa.groupby('positions').agg(
    jumlah_pemain=('name', 'count'),
    total_gaji_posisi=('wage_euro', 'sum'),
    rata_rata_rating=('overall_rating', 'mean')
).reset_index()

# 2. Filter posisi yang memiliki sampel pemain cukup banyak (misal minimal 10 pemain)
df_posisi = df_posisi[df_posisi['jumlah_pemain'] >= 10].copy()

# 3. Hitung Skor Efisiensi Gaji Posisi
# Rumus: Rata-rata Rating dibagi Total Gaji dalam ribuan Euro (agar angka proporsional)
df_posisi['skor_efisiensi_posisi'] = df_posisi['rata_rata_rating'] / (df_posisi['total_gaji_posisi'] / 1000)

# 4. Tampilkan Top 10 Posisi Bermain dengan Struktur Gaji Paling Sehat & Efisien
print("=== TOP 10 POSISI BERMAIN DENGAN STRUKTUR GAJI PALING EFISIEN ===")
print(df_posisi[['positions', 'jumlah_pemain', 'total_gaji_posisi', 'rata_rata_rating', 'skor_efisiensi_posisi']]
      .sort_values(by='skor_efisiensi_posisi', ascending=False)
      .head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

# Ambil 10 posisi teratas yang paling efisien gajinya
top_10_posisi = df_posisi.sort_values(by='skor_efisiensi_posisi', ascending=False).head(10)

plt.figure(figsize=(10, 6))

# Buat grafik batang
sns.barplot(
    x='skor_efisiensi_posisi',
    y='positions',
    hue='positions',
    data=top_10_posisi,
    palette='crest',
    legend=False
)

plt.title('Top 10 Posisi Bermain dengan Struktur Gaji Paling Sehat & Efisien', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Skor Efisiensi (Rata-rata Rating / Total Gaji dalam Ribuan Euro)', fontsize=12)
plt.ylabel('Posisi Bermain', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# --- INSIGHT 5: OPTIMALISASI JAM TERBANG RIIL BAKAT (PORSI PEMAIN MUDA BERDASARKAN POSISI) ---

# 1. Hitung total pemain per posisi bermain
df_total_posisi = df_fifa.groupby('positions').size().reset_index(name='total_pemain')

# 2. Filter dan hitung jumlah pemain muda (< 21 tahun) per posisi bermain
df_muda_posisi = df_fifa[df_fifa['age'] < 21].groupby('positions').size().reset_index(name='jumlah_pemain_muda')

# 3. Gabungkan datanya agar bisa dihitung persentasenya
df_jam_terbang = pd.merge(df_total_posisi, df_muda_posisi, on='positions', how='left')
df_jam_terbang['jumlah_pemain_muda'] = df_jam_terbang['jumlah_pemain_muda'].fillna(0)

# 4. Filter posisi yang memiliki sampel memadai (misal minimal 10 pemain secara keseluruhan)
df_jam_terbang = df_jam_terbang[df_jam_terbang['total_pemain'] >= 10].copy()

# 5. Hitung persentase porsi pemain muda pada masing-masing posisi
df_jam_terbang['persentase_pemain_muda'] = (df_jam_terbang['jumlah_pemain_muda'] / df_jam_terbang['total_pemain']) * 100

# 6. Tampilkan Top 10 Posisi yang paling banyak memberikan porsi untuk pemain muda akademi
print("=== TOP 10 POSISI DENGAN PORSI PEMAIN MUDA (<21 TAHUN) TERTINGGI ===")
print(df_jam_terbang[['positions', 'jumlah_pemain_muda', 'total_pemain', 'persentase_pemain_muda']]
      .sort_values(by='persentase_pemain_muda', ascending=False)
      .head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

# Ambil 10 posisi dengan persentase pemain muda tertinggi
top_10_muda = df_jam_terbang.sort_values(by='persentase_pemain_muda', ascending=False).head(10)

plt.figure(figsize=(10, 6))

# Buat grafik batang horizontal
sns.barplot(
    x='persentase_pemain_muda',
    y='positions',
    hue='positions',
    data=top_10_muda,
    palette='rocket', # Menggunakan palet gradasi warna merah-oranye yang tegas
    legend=False
)

plt.title('Top 10 Posisi Bermain yang Paling Banyak Memberikan Porsi Pemain Muda (<21 Tahun)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Persentase Pemain Muda dalam Posisi (%)', fontsize=12)
plt.ylabel('Posisi Bermain', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# --- INSIGHT 6: REKOMENDASI DAFTAR PEMAIN YANG WAJIB DIPINJAMKAN (LOAN TARGET) ---

# 1. Pastikan kolom growth_room sudah dihitung
df_fifa['growth_room'] = df_fifa['potential'] - df_fifa['overall_rating']

# 2. Filter Pemain sesuai Kriteria Matriks:
# - Usia antara 19 sampai 21 tahun
# - Overall Rating masih rendah (< 70) karena belum siap di skuad utama
# - Memiliki ruang pertumbuhan yang menjanjikan (Growth Room >= 10)
df_loan = df_fifa[
    (df_fifa['age'] >= 19) &
    (df_fifa['age'] <= 21) &
    (df_fifa['overall_rating'] < 70) &
    (df_fifa['growth_room'] >= 10)
].copy()

# 3. Urutkan berdasarkan sisa Growth Room tertinggi sebagai aset masa depan terpenting
print("=== TOP 10 PEMAIN USIA 19-21 TAHUN YANG WAJIB DIPINJAMKAN ===")
print(df_loan[['name', 'age', 'positions', 'overall_rating', 'potential', 'growth_room', 'value_euro']]
      .sort_values(by='growth_room', ascending=False)
      .head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

# Ambil data sampel top 50 pemain potensial usia 19-21 untuk visualisasi scatter
df_scatter_loan = df_fifa[(df_fifa['age'] >= 19) & (df_fifa['age'] <= 21)].sort_values(by='growth_room', ascending=False).head(50).copy()

# Buat kolom penanda apakah pemain direkomendasikan dipinjamkan atau tidak
df_scatter_loan['Status'] = np.where(df_scatter_loan['overall_rating'] < 70, 'Wajib Dipinjamkan', 'Keep di Tim Utama')

plt.figure(figsize=(10, 6))

# Buat Scatter Plot
sns.scatterplot(
    x='overall_rating',
    y='potential',
    hue='Status',
    style='Status',
    size='growth_room',
    sizes=(40, 240),
    palette={'Wajib Dipinjamkan': '#e74c3c', 'Keep di Tim Utama': '#2ecc71'}, # Merah untuk dipinjamkan, Hijau untuk aman
    data=df_scatter_loan
)

# Tambahkan garis batas rekomendasi di rating 70
plt.axvline(x=70, color='gray', linestyle='--', alpha=0.7, label='Batas Kesiapan Tim Utama (OVR 70)')

# Beri label nama pada 3 pemain teratas yang wajib dipinjamkan (paling kiri atas di zona merah)
df_top_loan = df_scatter_loan[df_scatter_loan['Status'] == 'Wajib Dipinjamkan'].head(3)
for index, row in df_top_loan.iterrows():
    plt.text(
        x=row['overall_rating'] + 0.5,
        y=row['potential'],
        s=row['name'],
        fontsize=9,
        fontweight='semibold'
    )

plt.title('Rekomendasi Pemain Usia 19-21 Tahun yang Wajib Dipinjamkan', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Overall Rating Saat Ini', fontsize=12)
plt.ylabel('Potential Rating', fontsize=12)
plt.legend(loc='lower right')

plt.tight_layout()
plt.show()
```

```python
# --- INSIGHT 7: IDENTIFIKASI KANDIDAT PENJUALAN SURPLUS ASET (ROI MAKSIMAL) ---

# 1. Pastikan tidak ada data kosong pada value dan wage
df_fifa['value_euro'] = df_fifa['value_euro'].fillna(0)
df_fifa['wage_euro'] = df_fifa['wage_euro'].fillna(0)

# 2. Hitung Rasio ROI (Nilai Pasar dibagi Beban Gaji Tahunan)
# Gaji mingguan dikali 52 untuk estimasi beban gaji dalam satu tahun
df_fifa['roi_ratio'] = df_fifa['value_euro'] / (df_fifa['wage_euro'] * 52)

# 3. Filter Pemain sesuai Kriteria Matriks:
# - Pemain yang sudah matang/bernilai tinggi (Overall Rating >= 75)
# - Memiliki nilai pasar nyata (Value > 0)
df_roi = df_fifa[(df_fifa['overall_rating'] >= 75) & (df_fifa['value_euro'] > 0)].copy()

# 4. Tampilkan Top 10 Pemain dengan Rasio ROI Penjualan Tertinggi (Surplus yang Siap Dijual)
print("=== TOP 10 KANDIDAT PENJUALAN UTAMA UNTUK MAKSIMALISASI ROI FINANSIAL ===")
print(df_roi[['name', 'age', 'positions', 'overall_rating', 'value_euro', 'wage_euro', 'roi_ratio']]
      .sort_values(by='roi_ratio', ascending=False)
      .head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

# Ambil 10 pemain dengan ROI rasio tertinggi
top_10_roi = df_roi.sort_values(by='roi_ratio', ascending=False).head(10)

plt.figure(figsize=(10, 6))

# Buat grafik batang horizontal
sns.barplot(
    x='roi_ratio',
    y='name',
    hue='name',
    data=top_10_roi,
    palette='YlOrBr_r', # Gradasi warna emas/oranye kecoklatan yang mewah
    legend=False
)

plt.title('Top 10 Kandidat Pemain dengan ROI Penjualan Tertinggi (Siap Jual)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Rasio Keuntungan (Nilai Pasar / Total Gaji Setahun)', fontsize=12)
plt.ylabel('Nama Pemain', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# --- INSIGHT 8: STANDARDISASI FILOSOFI TAKTIS (TACTICAL MATCH SCORE) ---

# 1. Tentukan atribut kunci untuk filosofi taktis (Contoh: Tiki-Taka / Attacking)
atribut_taktis = ['dribbling', 'ball_control', 'short_passing', 'acceleration']

# 2. Hitung rata-rata atribut ini untuk mendapatkan profil standar Tim Senior Utama (OVR >= 78)
profil_senior = df_fifa[df_fifa['overall_rating'] >= 78][atribut_taktis].mean()
print("=== PROFIL ATRIBUT STANDAR TIM SENIOR UTAMA ===")
print(profil_senior.round(1))
print("-" * 50)

# 3. Hitung Skor Kesesuaian Taktis (Tactical Match Score) untuk setiap pemain
# Rumusnya adalah rata-rata dari ke-4 atribut kunci tersebut
df_fifa['tactical_match_score'] = df_fifa[atribut_taktis].mean(axis=1)

# 4. Filter pemain muda (usia <= 21) yang memiliki kecocokan tinggi dengan filosofi senior
df_akademi_taktis = df_fifa[df_fifa['age'] <= 21].copy()

# 5. Tampilkan Top 10 Pemain Muda yang paling siap masuk ke sistem taktis Senior
print("=== TOP 10 PEMAIN MUDA DENGAN KESESUAIAN FILOSOFI TAKTIS TERTINGGI ===")
print(df_akademi_taktis[['name', 'age', 'positions', 'overall_rating', 'tactical_match_score']]
      .sort_values(by='tactical_match_score', ascending=False)
      .head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

plt.figure(figsize=(10, 6))

# Buat grafik distribusi kepadatan skor taktis pemain muda
sns.kdeplot(
    data=df_akademi_taktis,
    x='tactical_match_score',
    shade=True,
    color='#3498db',
    bw_adjust=1,
    label='Distribusi Pemain Muda (<21 Tahun)'
)

# Hitung nilai target 85% dari rata-rata standar senior untuk garis batas
target_skor_taktis = profil_senior.mean() * 0.85
plt.axvline(x=target_skor_taktis, color='#e74c3c', linestyle='--', linewidth=2,
            label=f'Target Filosofi Matriks (85% Senior: {target_skor_taktis:.1f})')

# Tambahkan teks penanda zona lolos filosofi
plt.text(target_skor_taktis + 2, plt.gca().get_ylim()[1] * 0.5, 'Zona Lolos\nFilosofi Taktis',
         color='green', fontsize=10, fontweight='bold')

plt.title('Distribusi Kesiapan Filosofi Taktis Pemain Muda vs Target Standar Senior', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Skor Kemampuan Taktis (Rata-rata Atribut Kunci)', fontsize=12)
plt.ylabel('Kepadatan Sampel (Density)', fontsize=12)
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# --- INSIGHT 9: MANAJEMEN RISIKO CEDERA ASET MUDA ---

# 1. Hitung Indeks Risiko Cedera (Injury Risk Index)
# Rumus: Akselerasi tinggi yang tidak diimbangi dengan Stamina & Kekuatan yang cukup
# Kita beri skala agar hasilnya berada di rentang yang mudah dibaca
df_fifa['injury_risk_index'] = (df_fifa['acceleration'] * 2) / (df_fifa['stamina'] + df_fifa['strength'])

# 2. Filter Pemain Akademi / Muda (Usia <= 21 tahun)
df_muda_fisik = df_fifa[df_fifa['age'] <= 21].copy()

# 3. Klasifikasikan Tingkat Risiko berdasarkan nilai indeks
def cek_kategori_risiko(val):
    if val > 1.2:
        return 'Risiko Tinggi (Wajib Rotasi)'
    elif val > 0.9:
        return 'Risiko Medium (Pantauan)'
    else:
        return 'Risiko Rendah (Aman)'

df_muda_fisik['kategori_risiko'] = df_muda_fisik['injury_risk_index'].apply(cek_kategori_risiko)

# 4. Tampilkan Top 10 Pemain Muda dengan Risiko Cedera Otot Tertinggi
print("=== TOP 10 PEMAIN MUDA DENGAN RISIKO CEDERA (OVERUSE) TERTINGGI ===")
print(df_muda_fisik[['name', 'age', 'positions', 'acceleration', 'stamina', 'strength', 'injury_risk_index', 'kategori_risiko']]
      .sort_values(by='injury_risk_index', ascending=False)
      .head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

plt.figure(figsize=(9, 5))

# Urutan kategori agar rapi dari risiko tertinggi ke terendah
order_risiko = ['Risiko Tinggi (Wajib Rotasi)', 'Risiko Medium (Pantauan)', 'Risiko Rendah (Aman)']

# Buat grafik count plot
sns.countplot(
    x='kategori_risiko',
    data=df_muda_fisik,
    order=order_risiko,
    palette={'Risiko Tinggi (Wajib Rotasi)': '#e74c3c',
             'Risiko Medium (Pantauan)': '#f39c12',
             'Risiko Rendah (Aman)': '#2ecc71'},
    hue='kategori_risiko',
    legend=False
)

# Tambahkan angka total pemain di atas setiap batang
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 8), textcoords='offset points', fontweight='bold')

plt.title('Peta Distribusi Risiko Cedera Pemain Muda Akademi (<21 Tahun)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Status Kategori Risiko Kebugaran', fontsize=12)
plt.ylabel('Jumlah Pemain (Orang)', fontsize=12)
plt.ylim(0, df_muda_fisik['kategori_risiko'].value_counts().max() * 1.15) # Beri ruang vertikal agar angka teks tidak terpotong

plt.tight_layout()
plt.show()
```

```python
# --- INSIGHT 10: PENINGKATAN DAYA TAWAR REPUTASI (ASET BRAND GLOBAL) ---

# 1. Filter Pemain Muda (Usia <= 21 tahun)
df_reputasi_muda = df_fifa[(df_fifa['age'] <= 21) & (df_fifa['potential'] >= 80)].copy()

# 2. Hitung Skor Daya Tawar (Leverage Score)
# Rumus: Menggabungkan nilai reputasi internasional saat ini dengan sisa potensi pertumbuhan mereka
df_reputasi_muda['leverage_score'] = df_reputasi_muda['international_reputation(1-5)'] * df_reputasi_muda['potential']

# 3. Tampilkan Top 10 Pemain Muda yang Menjadi Brand Ambassador Utama Akademi
print("=== TOP 10 TALENTA MUDA DENGAN DAYA TAWAR REPUTASI GLOBAL TERTINGGI ===")
print(df_reputasi_muda[['name', 'age', 'positions', 'international_reputation(1-5)', 'potential', 'leverage_score']]
      .sort_values(by='leverage_score', ascending=False)
      .head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

# Ambil 10 pemain muda dengan skor daya tawar tertinggi
top_10_leverage = df_reputasi_muda.sort_values(by='leverage_score', ascending=False).head(10)

plt.figure(figsize=(10, 6))

# Buat grafik batang horizontal
sns.barplot(
    x='leverage_score',
    y='name',
    hue='name',
    data=top_10_leverage,
    palette='viridis', # Gradasi warna hijau-biru-ungu yang modern dan bersih
    legend=False
)

plt.title('Top 10 Pemain Muda Akademi dengan Daya Tawar Reputasi Global Tertinggi', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Skor Daya Tawar (Reputasi Internasional x Potential Rating)', fontsize=12)
plt.ylabel('Nama Pemain', fontsize=12)

plt.tight_layout()
plt.show()
```
